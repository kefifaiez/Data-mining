```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Génération de données aléatoiresnp.random.seed
np.random.seed(40)
X = np.random.randn(30, 2)
X
```

```python

# Tracé du nuage de points initial
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nuage de points avant le groupement')
plt.show()
```

```python
# Création d'un objet KMeans avec k clusters
k=8
kmeans = KMeans(n_clusters=k)

# Entraînement du modèle
kmeans.fit(X)

# Obtention des affectations de cluster pour chaque échantillon
labels = kmeans.labels_

# Obtention des centres de cluster
centers = kmeans.cluster_centers_

#labels.shape
centers
```

```python
# Tracé du nuage de points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ({k} clusters)')
plt.show()
```

```python
# Liste pour stocker les valeurs d'inertie
inertie = []

# Nombre de clusters à tester
k_values = range(1, 15)

# Calcul de l'inertie pour chaque valeur de k
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertie.append(kmeans.inertia_)

# Tracé de la courbe du coude
plt.plot(k_values, inertie, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('Courbe du coude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Génération de données aléatoires
np.random.seed(42)
X = np.random.randn(30, 2)

# Valeurs de k à tester
k_values = [2, 4, 6, 10]

# Création de la figure avec 2x2 sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Parcours des différentes valeurs de k
for i, k in enumerate(k_values):
    # Création d'un objet KMeans avec k clusters
    kmeans = KMeans(n_clusters=k)
    # Entraînement du modèle
    kmeans.fit(X)
    # Obtention des affectations de cluster pour chaque échantillon
    labels = kmeans.labels_
     # Obtention des centres de cluster
    centers = kmeans.cluster_centers_
    # Tracé du nuage de points sur le sous-graphique correspondant
    ax = axes[i // 2, i % 2]
    ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    ax.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
    ax.set_title(f'k = {k}')

# Affichage de la figure
plt.tight_layout()
plt.show()
```
